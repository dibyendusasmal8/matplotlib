## Bar chart
- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups


```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='yellow')
```

```python
# horizontal bar chart
plt.barh(colors,children,color='green')
```

```python
# color and label
df = pd.read_csv('matplotlib_datasets1/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2016'],width=0.2)
plt.bar(df['batsman'],df['2015'],width=0.4)
plt.bar(df['batsman'],df['2017'],width=0.6)
```

```python
#Multi-Bar Plot for a Single Row
# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values1 = [5, 8, 12, 6]
values2 = [10, 6, 8, 15]

# Set up positions for bars
bar_width = 0.35
index = np.arange(len(categories))

# Create the multi-bar plot
plt.bar(index, values1, width=bar_width, label='Bar 1')
plt.bar(index + bar_width, values2, width=bar_width, label='Bar 2')

# Customize the plot
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Multi-Bar Plot for a Single Row')
plt.xticks(index + bar_width / 2, categories)
plt.legend()

# Show the plot
plt.show()
```

## Pie Chart
- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python

# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('matplotlib_datasets1/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
