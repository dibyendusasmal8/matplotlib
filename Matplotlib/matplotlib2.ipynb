# subplot

In Matplotlib, subplot is a function that allows you to create a grid of subplots within a single figure. It helps in organizing multiple plots in a single figure, making it easier to compare and analyze different data visualizations

```python
batter.head()
```

```python
# we know 
plt.scatter(batter['avg'],batter['strike_rate'])
plt.title("somting")
plt.xlabel('avg')
plt.ylabel('strike_rate')
```

```python
# subplot title/label
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(batter['avg'],batter['strike_rate'])
ax.set_title('avg vs srike ')
ax.set_xlabel('avg')
ax.set_ylabel('strike_rate')
```

```python
# subplot color/marker
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(batter['avg'],batter['strike_rate'],color='red',marker='+')
ax.set_title('avg vs srike ')
ax.set_xlabel('avg')
ax.set_ylabel('strike_rate')
fig.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,figsize=(10,7))
ax[0].scatter(batter['avg'],batter['strike_rate'],color = 'red')
ax[1].scatter(batter['avg'],batter['runs'],marker='+')

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

## 3D Surface Plots

```python
# x and y are 1D arrays representing the coordinates along the x and y axes, respectively.
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
# np.meshgrid(x, y) creates two 2D arrays, XX and YY, where each element of Xcorresponds to a specific x-coordinate and each element of YY corresponds to a specific y-coordinate.
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# Create 1D arrays for x and y coordinates
x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)

# Create coordinate matrices using meshgrid
X, Y = np.meshgrid(x, y)

# X and Y are now 2D arrays representing the coordinates of each point in the grid

# Example: Evaluate a function over the grid (e.g., the Gaussian function)
Z = np.exp(-(X**2 + Y**2) / 10)

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

plt.show()
```

```python
xx,yy = np.meshgrid(x,y)

z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## contour 
- this funtion make a 3d plot into a 2d plot

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## contourf
- this funtion make 3d into a fild 2d plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
